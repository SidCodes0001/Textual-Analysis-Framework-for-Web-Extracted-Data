```python
import requests
from bs4 import BeautifulSoup
import openpyxl
import os
import nltk
```

```python
def extract_text(url, url_id):
  
  # this function extracts article title and text from a URL and saves it to a text file named after url_id.


  try:
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for bad status codes
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find elements containing title and text (adjust selectors based on website structure)
    title_element = soup.find('title').string  # Replace with appropriate selector
    # text_elements = soup.find_all('p', class_='article-body')  # Replace with appropriate selector
    text_elements = soup.find_all('div', class_="td-post-content")


    # Extracting and clean text using .text and then using .strip func followed by .lower for symmetry across the entire document and it will be easier for further analysis 
    title = title_element.text.strip() if title_element else "" # if unable to extract anything the we put nothing in document
    title = title + '.' # adding this because , considering title as seperate line while doing sentence tokenization
    title=title.lower()
    text = "\n".join([p.text.strip() for p in text_elements])
    text=text.lower()

    # Save extracted text to file
    with open(f"output/{url_id}.txt", "w", encoding="utf-8") as f:
      f.write(f"{title}\n\n{text}")
    print(f"Extracted text from {url} and saved to {url_id}.txt")
  except requests.exceptions.RequestException as e:
    print(f"Error fetching URL {url}: {e}")


```

```python

# Reads URLs from "Input.xlsx" and calls extract_text for each.
  
wb = openpyxl.load_workbook("Input.xlsx")
sheet = wb.active

  # URL_ID is in the first column and URL is in the second
for row in sheet.iter_rows(min_row=2):
    url_id = row[0].value
    url = row[1].value
    extract_text(url, url_id)

```

# Combining Stopwords in a single file

```python
# Merging all stopwords file into one 

def combine_stopwords(stopword_files, output_file):
  # i have used set for this task because in set we dont have repeated elements 
  combined_stopwords = set()  

  for stopword_file in stopword_files:
    try:
      with open(stopword_file, "r") as f:
        combined_stopwords.update([line.strip().lower() for line in f])
    except Exception as e:
      print(f"Error: Stopword file not found: {e}")

  with open(output_file, "w", encoding="utf-8") as f:
    f.write("\n".join(combined_stopwords))

# we can generate this list using list.dir and os.joi but since there were few of them so i prefered the manual way 
stopword_files = ["stopwords\StopWords_Auditor.txt", "stopwords\StopWords_Currencies.txt","stopwords\StopWords_DatesandNumbers.txt","stopwords\StopWords_Generic.txt","stopwords\StopWords_GenericLong.txt","stopwords\StopWords_Geographic.txt","stopwords\StopWords_Names.txt"]
output_file = "combined_stopwords.txt"
combine_stopwords(stopword_files, output_file)
print(f"Combined stopwords saved to: {output_file}")
```

### Listing all the files in output folder(in this we have all the articles data). We need this to perform stop word removal in next step


```python
import os
file_list = os.listdir('output')

final_file_list = []

base_path = 'output/'

for single_file_path in file_list:
    final_file_list.append(os.path.join(base_path,single_file_path))


# file_list = ['output/bctech2011.txt']

print(file_list)
print(final_file_list)
```

```python
import os

# this function takes three parameters as input file list in which we have all the location of all the files , list of stopwords, and path of output folder in which it will store the filtered files 

def remove_stopwords(file_list, stopwords, output_folder):
  
  # file list ------> have all path to individual articles 
  # file path will be individual path to a specific article file named after url_id
  
  for file_path in file_list:
    try:
      with open(file_path, "r", encoding="utf-8") as f:
        text = f.read()
    
      text = text.lower()
      text_list = [word for word in text.split() if word not in stopwords]
      filtered_text = " ".join(text_list) # we are doing this to make entire text a single element , so it will be easier for further analysis.

      # Create output filename with same name as url_id
      #split func returns two tuple  ----> one with file name and other with extension

      output_filename = os.path.splitext(os.path.basename(file_path))[0] + ".txt"  # this will return ---> "filename" + ".txt"
      output_path = os.path.join(output_folder, output_filename)

      with open(output_path, "w", encoding="utf-8") as f:
        f.write(filtered_text)
      print(f"Stopwords removed and saved to: {output_path}")
    except FileNotFoundError:
      print(f"Error: File not found: {file_path}")

output_folder = "filteredFiles"

# Creating output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

remove_stopwords(final_file_list, stopwords_list, output_folder)
```

# Till now we have 
1) master list for positive and negative words,
2) a combined stopwords file and list
3) output files with title and article content
4) filteredFiles in which we have removed stop words 


Now we have to do is sentence aand word tokenize the filteredFiles for further processing and calculating the desired variables.

```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
import re
import os
import openpyxl


# =================================================================== iterates through each file in filtered file list =================================================================================

# positive_words_list ----------> list of all positive words extracted from given file 
# negative_words_list ----------> list of all negative words extracted from given file 

# we have removed stopwords using combined_stopwords file but for word count we also need to remove punctuations 

base_path_for_filteredFiles = 'filteredFiles'

for i in range (len(os.listdir('filteredFiles'))): # this loop will run for no of files in filteredFiles dir i.e. total no of files times 
    #getting the file path to do tasks 
    file_path_for_variable_calculations = (os.path.join(base_path_for_filteredFiles,os.listdir('filteredFiles')[i]))

    file_name_url_id = os.path.splitext(os.path.basename(os.listdir('filteredFiles')[i]))[0]

    print(file_name_url_id)

    #declaring variables 
    positive_score = 0
    negative_score = 0

    complex_word_count = 0

    vowels = "aeiouyAEIOUY"
    exceptions = ("es", "ed")
    # need this to calculate average word length
    total_characters = 0 

    complex_words_list = []

    pronouns = ["I", "me", "my", "mine", "myself", "we", "us", "our", "ours", "ourselves","you", "your", "yours", "yourself", "yourselves", "he", "him", "his", "himself","she", "her", "hers", "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves"]

    no_of_pronouns =0


    #opening file path using base value and i path 
    with open(file_path_for_variable_calculations , "r",encoding='utf-8') as f:  # i have used utf encoding because , some files were giving error while executing code
        text = f.read()
        sentences = sent_tokenize(text)
        # words = word_tokenize(text)
        words = re.findall(r"\w+", text) # this focuses only on alpha-numeric words only, i have used regex because for word tokenize nltk in some cases doesn't perform good
        # words = list(set(words)) # i am assuming while coding that we need to consider on only unique words 

    # for calculating positive words , negative words and no of complex words 
    for word in words:
        vowel_count = sum(w in vowels for w in word) 

        # Handle exceptions for "es" and "ed" endings
        if vowel_count <= 2 and word.endswith(exceptions):
          vowel_count -= 1

        if vowel_count > 2:
          complex_word_count += 1
          complex_words_list.append(word)
        #   complex_words_list = list(set(complex_words_list)
        
        #calculating positive and negative score
        if word in positive_words_list:
            positive_score = positive_score+ 1
        elif word in negative_words_list:
            negative_score = negative_score+ 1


        #code for calculating Average Word Length
        total_characters += len(word)

        #calculating pronouns 
        if word in pronouns:
            no_of_pronouns+=1
    
        
# =================================================================== calculating respective values for each file  =================================================================================

    total_words_in_doc = len(words)
    print(total_words_in_doc)
    total_sentences_in_doc = len(sentences)

# complex_words_list = list(set(complex_words_list))
    total_complex_words = len(complex_words_list)

    polarity_score = (positive_score-negative_score) / ((positive_score+negative_score)+0.000001)

    subjectivity_score = (positive_score+negative_score) / ((total_words_in_doc)+0.000001)

    average_sentence_length = int(total_words_in_doc/total_sentences_in_doc)

    percentage_of_complex_words = total_complex_words / total_words_in_doc

    fog_index = 0.4 * (average_sentence_length + percentage_of_complex_words)

    average_word_length = int(total_characters / total_words_in_doc)

    '''
    Now we have the following things : 

    1. Positive Score
    2. Neagtive Score
    3. Polarity Score
    4. SUBJECTIVITY SCORE
    5. AVG SENTENCE LENGTH = AVG NUMBER OF WORDS PER SENTENCE -------> that's my opinion 
    6. Percentage of complex words
    7. Fog Index
    8. Complex word count
    9. Word count
    10. Syllable per word -----> i have doubt in this thing, but i have used this to count no of complex words
    11. personal pronoun
    12. Average word length 

    '''

    # =================================================================== putting values in output file   =================================================================================

    wb = openpyxl.load_workbook("output_file.xlsx")
    sheet = wb.active

    for row in sheet.iter_rows(min_row=2):
       if row[0].value.strip() == file_name_url_id.strip():
            row[2].value = positive_score
            row[3].value = negative_score
            row[4].value = polarity_score
            row[5].value = subjectivity_score
            row[6].value = average_sentence_length
            row[7].value = percentage_of_complex_words
            row[8].value = fog_index
            row[9].value = average_sentence_length
            row[10].value = total_complex_words
            row[11].value = total_words_in_doc
            # row[12].value = -------> syllabe per word...
            row[13].value = no_of_pronouns
            row[14].value = average_word_length
            
    wb.save("output_file.xlsx")

          


```
